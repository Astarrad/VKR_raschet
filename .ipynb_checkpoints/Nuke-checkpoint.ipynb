```python
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
import seaborn as sns
```

## Приближенный расчет параметров ядерной энергетической установки с встроенным в активную зону термоэмиссионным преобразователем одноэлементной конструкции

Исходные данные:

```python
P_el = 100*10**3 # Вт - требуемая электрическая мощность ЯЭУ
t = 7 # лет - требуемый ресурс работы ЯЭУ
d_k = 0.02 #
L_k = 0.25 #
ro_t = 19050 #
ro_z = 5600 #
ro_otr = 1950 #
delta = 0.06 #
ro_izl = 8920 #
ro_lih = 2900 # кг/м^3 - плотность гидрида лития
ro_pb = 11340
ro_al = 2800
d_z = 0.002 # м - толщина алюминиевого корпуса защиты
```

 1. Суммарная поверхность энерговыделяющих катодов $ S_к, см^2,$ определяется по эллектрической мощности ЯЭУ $P_{эл}, Вт: $

Для выбора $P_{эл} и P_{уд}$ используются эмпирические данные, из таблиц 1 и 2. Для определения этих параметров следует пролонгировать исходные соотношения методами линейной эктраполяции:

```python
t_i = np.arange(1,11,1)
k = -125
b = 2150
def T_k(z):
    return k*z+b # Линейная функция зависимости температуры катода от времени ресурса работы ЯЭУ,
                 # выраженная на основе графика 1
```

Отрисуем график зависимости ресурса работы ядерной энергетической установки от температуры катода, и отыщем температуру для искомого ресурса - $T_k(7)$:

```python
lplot = sns.lineplot(T_k(t_i), t_i)

lplot.set_title('Зависимость ресурса работы ядерной энергетической установки от температуры катода')
lplot.set_xlabel('T_k')
lplot.set_ylabel('t')
T_k = T_k(t)
print(f'T_k({t}) =',T_k)
```

Далее требуется экстраполировать график зависимости удельного энергосъёма от температуры катода

```python
# Экстраполяция с помощью линейной функции
points = np.arange(1500, 2050, 50)
P_ud_i = np.array([2,2.117647,2.195294,2.411764,2.705882,3,3.235294,3.823529,4.294117,4.647058,5])
n = points.size
```

```python
nx = 3 # "nx-1" - количество точек аппроксимации. В ходе рассчетов было принято аппроксимировать
      # кусочную функцию по 2 точкам

res = scipy.stats.linregress(points[0:nx], P_ud_i[0:nx]) # Вычислим параметры линейной функции по k точкам.

# Построим графики: исходный и кусочно-линейную аппроксимацию требуемого диапазона значений
Pud_Tk = sns.lineplot(points, P_ud_i, label='Исходный график', sizes=[1000,2000])
Pud_Tk = sns.lineplot(points, res.intercept + res.slope*points, label='Аппроксимированная линия', sizes=[1000,2000])
Pud_Tk.set_title('Зависимость удельного энергосъема от температуры катода')
Pud_Tk.set_xlabel('T_к')
Pud_Tk.set_ylabel('P_уд')

# Отыскание P_уд в искомой точке
P_ud = res.intercept + res.slope*T_k
print(f'P_уд({T_k}) =', P_ud)
```

```python
S_k = (1.2 * P_el)/(P_ud*10000)
S_k
```

Далее необходимо экстраполировать график зависимости КПД ЯЭУ от температуры катода. График имеет сложную форму.
В ходе расчетов было предпринято несколько попыток экстраполяции с различными целевыми функциями: линейной, показательной, степенной.

```python
def func(x, a, b, c):
    return (x+b)**a+c
```

```python
ydata3 = np.array([0.05, 0.065, 0.07739, 0.08, 0.08923, 0.09538, 0.1, 0.10769, 0.11384, 0.11538, 0.12])
popt, pcov = scipy.optimize.curve_fit(func, points, ydata3)
```

```python
res2 = scipy.stats.linregress(points[4:], ydata3[4:])
plt.plot(points, ydata3)
plt.plot(points, res2.intercept + res2.slope*points)
Eta_lin = res2.intercept + res2.slope*T_k
Eta_lin
```

```python
plt.plot(points, func(points, *popt), 'r-',
         label='fit: a=%5.3f, b=%5.3f, c=%5.3f' % tuple(popt))
plt.plot(points, ydata3)
```

```python
be = np.arange(1000,3000,10)
KPD_Tk = sns.lineplot(be, func(be, *popt), label='Экстраполяция степенной функцией')
KPD_Tk = sns.lineplot(points, ydata3, label='Исходный график')
KPD_Tk = sns.lineplot(be, res2.intercept + res2.slope*be, label='Линейная экстраполяция')
KPD_Tk.set_xlabel('T_к')
KPD_Tk.set_ylabel('КПД')
Eta_st = func(T_k,*popt)
Eta_st
```

Как можно увидеть, линейная экстраполяция лучше подходит для последующих расчетов, так как экстраполяция степенной функцией не способна выразить реальный результат.

Cледующие расчеты будут выполняться с допущением, что зависимость КПД ЯЭУ от температуры катода - линейна.

2. Эмисионная поверхность катода.

```python
S_egk = np.pi * d_k * L_k
S_egk
```

3. Число каналов ЭГК.

```python
N_k = S_k / S_egk
N_k = math.ceil(N_k)
N_k
```

4. Относительный шаг решетки и значение пористости активной зоны.

```python
PSI = 40/20
PSI
```

```python
KSI = 1-np.pi/(2*3**(1/2)*PSI**2)
KSI
```

5. Диаметр активной зоны.

```python
D_a = np.sqrt((d_k**2*N_k)/(1-KSI))
D_a
```

6. Объём активной зоны.

```python
V_a = np.pi*D_a**2*L_k/4
V_a
```

7. Длину активной зоны принимаем равной длине ЭГК.

```python
L = L_k
```

8. Масса активной зоны. Масса топлива и масса замедлителя.

```python
G_t = N_k*np.pi*d_k**2*L*ro_t/4
G_t
```

```python
G_z = (np.pi*L/4)*(D_a**2-d_k**2*N_k)*ro_z
G_z
```

9. Масса отражателя.

```python
G_otr = (2*np.pi*(D_a+2*delta)**2)/4*(delta*ro_otr + np.pi*(D_a+2*delta)*L*delta*ro_otr)
G_otr
```

10. Диаметр реактора.

```python
D_r = D_a + 2*delta
D_r
```

11. Масса реактора. Сумма масс топлива, замедлителя, отражателя.

```python
G_r = G_t + G_z + G_otr
G_r
```

12. Тепловая мощность установки, с учетом электрической мощности и КПД преобразования теплоты в электрическую энергию.

```python
Q_t = (1.2 * P_el) / Eta_lin
Q_t
```

13. Количество теплоты сбрасываемой с излучателя.

```python
Q_izl = Q_t - 1.2*P_el
Q_izl
```

14. Площадь поверхности излучателя.

```python
eps = 0.88
sig = 5.67*10**-8
T = 800
F_izl = Q_izl/(eps*sig*T**4)
F_izl
```

15. Масса излучателя.

```python
d_izl = 0.09
G_izl = F_izl*d_izl*ro_izl
G_izl
```

16. Доза стойкости аппаратуры.

```python
Doz_a = 10**16
```

17. Доза излучения реактора.

```python
F_az = np.pi*D_a**2/4
F_az
```

```python
Doz_r = 7.8*10**8*Q_t*t/F_az
Doz_r
```

18. Толщина слоя лития.

```python
B = 1
t_z = np.log(Doz_a/(B*Doz_r))/-0.114
t_z
```

19. Толщина слоя свинца.

```python
def func(x, a, b):
    return a*np.log(x)+b
```

```python
B2 = np.array([10,5.72727,3.16667,1.84091,1])
t_pb_i = np.array([0,2,4,6,8])
popt, pcov = scipy.optimize.curve_fit(func, B2, t_pb_i)
```

```python
X = sns.lineplot(func(range(11), *popt),range(11))
```

```python
t_pb = func(B, *popt)
t_pb = t_pb/1000
t_pb
```

20. Площадь поперечного сечения реактора.

```python
F_r = np.pi*(D_a+2*delta)**2/4
F_r
```

21. Объём защиты.

```python
V_z = F_r * t_z
V_z
```

22. Масса гидрида лития.

```python
G_lih = V_z * ro_lih
G_lih
```

23. Объём свинца.

```python
V_pb = F_r * t_pb
V_pb
```

24. Масса свинца.

```python
G_pb = V_pb * ro_pb
G_pb
```

25. Масса оболочки

```python
G_ob = 2 * F_r * d_z * ro_al + np.pi * (D_a + 2 *delta) * t_z * ro_al * d_z
G_ob
```

26. Общая масса реактора-генератора

```python
G_rg = G_ob + G_lih + G_pb + G_r + G_izl
G_rg
```
